```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import os
import random
import json
import sqlite3
import mysql.connector
# === CONFIG ===
IMAGE_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 1
DATASET_DIR = 'D:\\adversarial\\dataset'  # Folder containing all class subfolders
DB_PATH = 'traffic_signals.db'

# === 1. Load Image Paths and Labels ===
image_paths = []
labels = []
valid_extensions = ('.jpg', '.jpeg', '.png')

class_names = sorted([
    d for d in os.listdir(DATASET_DIR)
    if os.path.isdir(os.path.join(DATASET_DIR, d)) and not d.startswith('.')
])
print(class_names)

# === 1.1 Connect to SQLite and create table ===
conn = mysql.connector.connect(user="AJ1070", password="Aloysius@107", host="dataset1.mysql.database.azure.com", port=3306, database="dataset1", ssl_ca="./DigiCertGlobalRootCA.crt.pem", ssl_disabled=False)
```

```python


# === 1.2 Insert image info into DB and list for training ===
for label_index, class_name in enumerate(tqdm(class_names)):
    class_folder = os.path.join(DATASET_DIR, class_name)
    for filename in tqdm(os.listdir(class_folder),leave=False):
        if filename.lower().endswith(valid_extensions):
            image_path = os.path.join(class_folder, filename)
            image_paths.append(image_path)
            labels.append(label_index)
            
            cursor.execute('''
    INSERT INTO traffic_signals (filename, label, filepath)
    VALUES (%s, %s, %s)
''', (filename, class_name, image_path))

conn.commit()
```

```python



# === 2. Split into Train and Test Sets ===
train_paths, test_paths, train_labels, test_labels = train_test_split(
    image_paths, labels, test_size=0.2, stratify=labels, random_state=42)
import tensorflow as tf
# === 3. Function to Load and Preprocess Image ===
def preprocess_image(file_path, label):
    image = tf.io.read_file(file_path)
    image = tf.image.decode_image(image, channels=3, expand_animations=False)
    image = tf.image.resize(image, IMAGE_SIZE)
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# === 4. Create TensorFlow Datasets ===
train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_ds = train_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))
test_ds = test_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# === 5. Build the Model ===
num_classes = len(class_names)

model = tf.keras.Sequential([
    tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# === 6. Compile and Train ===
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_ds, validation_data=test_ds, epochs=EPOCHS)

# === 7. Evaluate on Test Set ===
test_loss, test_accuracy = model.evaluate(test_ds)
print(f"\n✅ Test Accuracy: {test_accuracy:.2f}")

# === 8. Save Model & Class Names ===
model.save("model.h5")
with open("class_names.json", "w") as f:
    json.dump(class_names, f)

# === 9. Predict on a New Sample ===
def predict_image(image_path):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, IMAGE_SIZE)
    img = tf.expand_dims(img, axis=0) / 255.0

    predictions = model.predict(img)
    predicted_index = np.argmax(predictions)
    predicted_class = class_names[predicted_index]

    # Show image with predicted label
    plt.imshow(plt.imread(image_path))
    plt.title(f"Predicted: {predicted_class}")
    plt.axis('off')
    plt.show()

    return predicted_class

# === 10. Example Usage ===
sample_image = random.choice(image_paths)
print("\n🔍 Sample image path:", sample_image)
predicted = predict_image(sample_image)
print("📌 Predicted label:", predicted)
```

```python
import os
import cv2
import numpy as np
import pandas as pd
import sqlite3

# === STEP 1: Generate image-label DataFrame ===
def label_images_from_directory(dataset_dir):
    data = []
    class_names = sorted(os.listdir(dataset_dir))
    class_indices = {name: idx for idx, name in enumerate(class_names)}

    for class_name in class_names:
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):
            for filename in os.listdir(class_path):
                if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(class_path, filename)
                    label = class_indices[class_name]
                    data.append((image_path, class_name, label))

    return pd.DataFrame(data, columns=["image_path", "class_name", "label"])

# === STEP 2: Load base image ===
base_image_path = 'stop.png'
base_image = cv2.imread(base_image_path, cv2.IMREAD_GRAYSCALE)

if base_image is None:
    raise FileNotFoundError(f"Could not read base image: {base_image_path}")

scale_factor = 0.7
base_image = cv2.resize(
    base_image,
    (int(base_image.shape[1] * scale_factor), int(base_image.shape[0] * scale_factor)),
    interpolation=cv2.INTER_AREA
)

# === STEP 3: Initialize ORB detector ===
orb = cv2.ORB_create(nfeatures=1000)
base_kp, base_des = orb.detectAndCompute(base_image, None)

# === STEP 4: Load dataset ===
dataset_dir = 'D:\\adversarial\\dataset'
image_labels_df = label_images_from_directory(dataset_dir)

# === STEP 5: Set up matcher ===
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

best_score = 0
best_match_row = None
best_match_img = None

# === STEP 6: Feature Matching ===
for idx, row in image_labels_df.iterrows():
    candidate_path = row["image_path"]
    candidate_img = cv2.imread(candidate_path, cv2.IMREAD_GRAYSCALE)

    if candidate_img is None:
        print(f"⚠️ Could not read image: {candidate_path}")
        continue

    try:
        kp, des = orb.detectAndCompute(candidate_img, None)
        if des is None:
            continue

        matches = bf.match(base_des, des)
        matches = sorted(matches, key=lambda x: x.distance)

        good_matches = [m for m in matches if m.distance < 50]
        score = len(good_matches)

        if score > best_score:
            best_score = score
            best_match_row = row
            best_match_img = candidate_img

    except Exception as e:
        print(f"❌ Error with {candidate_path}: {e}")
        continue

# === STEP 7: Output ===
if best_match_row is not None:
    print("\n✅ Best Match Found!")
    print(f"🖼️  Image Path: {best_match_row['image_path']}")
    print(f"🏷️  Class Name: {best_match_row['class_name']}")
    print(f"🔢 Label Index: {best_match_row['label']}")
    print(f"✨ Good Matches: {best_score}")

    # === INSERT INTO DATABASE ===
    db_path = 'traffic_signals.db'
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    base_filename = os.path.basename(base_image_path)
    matched_label = best_match_row['class_name']

    cursor.execute('''
        INSERT INTO traffic_signals (filename, label, filepath)
        VALUES (?, ?, ?)
    ''', (base_filename, matched_label, base_image_path))

    conn.commit()
    conn.close()

    print(f"🗃️ Stored '{base_filename}' in DB with label: {matched_label}")

    # === Resize candidate to match base image size ===
    scale_x = base_image.shape[1] / best_match_img.shape[1]
    scale_y = base_image.shape[0] / best_match_img.shape[0]
    resize_factor = (scale_x + scale_y) / 2.0

    resized_candidate_img = cv2.resize(
        best_match_img,
        (int(best_match_img.shape[1] * resize_factor), int(best_match_img.shape[0] * resize_factor)),
        interpolation=cv2.INTER_AREA
    )

    kp2, des2 = orb.detectAndCompute(resized_candidate_img, None)
    matches = bf.match(base_des, des2)
    matches = sorted(matches, key=lambda x: x.distance)

    match_vis = cv2.drawMatches(base_image, base_kp, resized_candidate_img, kp2, matches[:20], None, flags=2)

    cv2.imshow("Best Match (Resized to Match Template)", match_vis)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("❌ No suitable match found.")
    match_vis = None

# === Save the match result image ===
if match_vis is not None:
    output_match_path = "best_match_result.jpg"
    cv2.imwrite(output_match_path, match_vis)
    print(f"💾 Match image saved to: {output_match_path}")

# === Save match details ===
if best_match_row is not None:
    match_details_path = "match_details.csv"
    pd.DataFrame([best_match_row]).to_csv(match_details_path, index=False)
    print(f"📄 Match details saved to: {match_details_path}")
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# === STEP 6: Generate Adversarial Perturbation on Best Match ===

def load_and_preprocess_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3, expand_animations=False)
    image = tf.image.resize(image, IMAGE_SIZE)
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.expand_dims(image, axis=0)
    return image

def convert_to_one_hot(label, num_classes):
    return tf.one_hot(label, num_classes)

loss_object = tf.keras.losses.CategoricalCrossentropy()

# Targeted FGSM (to force a specific label)
def create_targeted_adversarial_pattern(input_image, target_label, model):
    with tf.GradientTape() as tape:
        tape.watch(input_image)
        prediction = model(input_image, training=False)
        loss = loss_object(target_label, prediction)
    gradient = tape.gradient(loss, input_image)
    signed_grad = -tf.sign(gradient)  # Negate for targeted FGSM
    return signed_grad

# === Start if a match is found ===
if best_match_row is not None:
    print("\n🔍 Generating targeted adversarial perturbation...")

    # Load original image and label
    input_image = load_and_preprocess_image(best_match_row['image_path'])
    original_label = best_match_row['label']
    input_label = convert_to_one_hot(original_label, num_classes=num_classes)
    input_label = tf.expand_dims(input_label, axis=0)

    # Get original prediction
    original_pred = model.predict(input_image)
    original_class_index = np.argmax(original_pred)
    original_confidence = np.max(original_pred)
    original_class_name = class_names[original_class_index]

    print(f"✅ Original prediction: {original_class_name} ({original_confidence:.2f})")

    # === Pick a different target class to attack toward
    target_class_index = (original_class_index + 5) % num_classes  # just for example
    target_label = convert_to_one_hot(target_class_index, num_classes)
    target_label = tf.expand_dims(target_label, axis=0)
    target_class_name = class_names[target_class_index]
    print(target_class_name)

    print(f"🎯 Targeted misclassification: {target_class_name}")

    # Create adversarial pattern to trick model into predicting target class
    perturbations = create_targeted_adversarial_pattern(input_image, target_label, model)

    # Show perturbation
    plt.figure(figsize=(5, 5))
    plt.imshow(perturbations[0] * 0.5 + 0.5)
    plt.title("⚠️ Targeted Adversarial Perturbation")
    plt.axis('off')
    plt.show()

    # Apply perturbation
    epsilon = 0.25  # aggressive perturbation
    adversarial_image = input_image + epsilon * perturbations
    adversarial_image = tf.clip_by_value(adversarial_image, 0.0, 1.0)


    # Predict on adversarial image
    adv_pred = model.predict(adversarial_image)
    adv_class_index = np.argmax(adv_pred)
    adv_confidence = np.max(adv_pred)
    adv_class_name = class_names[adv_class_index]

    print(f"⚠️ Adversarial prediction: {adv_class_name} ({adv_confidence:.2f})")

    # Show adversarial image
    plt.figure(figsize=(5, 5))
    plt.imshow(adversarial_image[0])
    plt.title(f"🔥 Adversarial Image: {adv_class_name} ({adv_confidence:.2f})")
    plt.axis('off')
    plt.show()

    # Merge: mostly adversarial
    merged_image = (0.3 * input_image + 0.7 * adversarial_image)
    merged_image = tf.clip_by_value(merged_image, 0.0, 1.0)

    # Predict on merged image
    merged_pred = model.predict(merged_image)
    merged_class_index = np.argmax(merged_pred)
    merged_confidence = np.max(merged_pred)
    merged_class_name = target_class_name

    # Show merged image
    plt.figure(figsize=(5, 5))
    plt.imshow(merged_image[0])
    plt.title(f"🌀 Merged Image: {merged_class_name} ({merged_confidence:.2f})")
    plt.axis('off')
    plt.show()

    print(f"\n🔀 After merging, the label changed to: {merged_class_name} with confidence: {merged_confidence:.2f}")

    # Confidence change report
    print(f"📉 Confidence dropped from {original_class_name} ({original_confidence:.2f}) → {merged_class_name} ({merged_confidence:.2f})")

else:
    print("⚠️ Cannot generate adversarial image without a matched image.")

# Save the adversarial image
adv_image_path = "adversarial_image.jpg"
adv_np = (adversarial_image[0].numpy() * 255).astype(np.uint8)
tf.keras.preprocessing.image.save_img(adv_image_path, adv_np)
print(f"💾 Adversarial image saved as: {adv_image_path}")

with open("adversarial_info.txt", "w") as f:
    f.write(f"{merged_class_name},{merged_confidence:.4f}")



```
